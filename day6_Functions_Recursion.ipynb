# Recursion

When a function calls itself repeatedly.

```python
# prints n to 1 backwards

def show(n):
    if(n==0):  #base case
        return
    print(n,end="\t")
    show(n-1)
    print("End")

show(5)
```

```python
def fact(n):
    if (n ==0 or n==1):
        return 1
    else:
        return n * fact(n-1)

fact(5)
```

```python
# Write a recursive function to calculate the sum of first n natural numbers.

def sum(n):
    if (n==0):
        return 0
    else:
        return n + sum(n-1)

sum(3)
```

```python
# Write a recursive function to print all elements in a list.

def print_list(lst, index=0):
    # Base case: stop when index reaches length of list
    if index == len(lst):
        return

    print(lst[index])
    print_list(lst, index + 1)

my_list = [10, 20, 30, 40, 50]
print_list(my_list)
```
