# Operators
* Arithmatic operators (+,-,*,/,%,**)
* Relational / Comparsion operators (==,!=,>,<,>=,<=)
* Assignment Operators (=,+=,-=,*=,%=,**=)
* Logical Operators (not,and,or)

```python
#Arithmatic operators

a = 9
b = 2

print (a+b)
print (a-b)
print (a*b)
print (a/b)
print (a%b)
print (a**b)
```

```python
#Relational / Comparsion operators

a = 50
b = 30

print (a==b)
print (a!=b)
print (a>b)
print (a<b)
print (a>=b)
print (a<=b)
```

```python
#Assignment Operators

a = 20
b = 30

a += b
print(a)

a = 20 
a -= b
print(a) 

a = 20 
a *= b
print(a) 

a = 20 
a /= b
print(a) 

a = 20 
a %= b
print(a) 

a = 20 
a **= b
print(a) 
```

```python
#Logical Operators 

a = 20
b = 30
value1 = True
value2 = False

print (not False)
print (not (a>b))
print ("Value are" , value1 and value2)
print ((a==b) or (a>b))
```
