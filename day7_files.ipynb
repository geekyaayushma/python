# File I/O


There are two different methods for opening a file :

* "r" - Read - Default value. Opens a file for reading, error if the file does not exist

* "w" - Write - Opens a file for writing, creates the file if it does not exist

* "a" - Append - Opens a file for appending, creates the file if it does not exist

* "x" - create a new file and open it for writing

* "b" - binary - binary mode

* "t" - text - open text file (default)

* "+" - open a disk file for updating (reading and writing)

```python
import os

os.makedirs("/kaggle/working/file2io", exist_ok=True)

with open("/kaggle/working/file2io/file2.txt", "w") as f:
    f.write("I want to learn javascript")

with open("/kaggle/working/file2io/file2.txt", "r") as f:
    print(f.read())
```

```python
with open("/kaggle/working/file2io/file2.txt", "a") as f:
    f.write("\n I'm learning python")

with open("/kaggle/working/file2io/file2.txt", "r") as f:
    print(f.read())
```

```python
with open("/kaggle/working/file2io/file2.txt", "r+") as f:
    f.write("\n I'm learning python")
    f.seek(0)        # go back to start
    print(f.read())  # now reads the whole file

# r+ point/write at starting of file
```

```python
with open("/kaggle/working/file2io/file2.txt", "w+") as f:
    f.write("I'm learning python")
    f.seek(0)
    print(f.read())

#w+ truncates/clear the file and write
```

```python
with open("/kaggle/working/file2io/file2.txt", "a+") as f:
    f.write("I'm learning javascript too")
    f.seek(0)
    print(f.read())

# pointer writes at end
```

# Practice

```python
# create a new file "practice.txt" . Add th following data in it:
# Hi everyone
# we are learning file i/o
# using java
# I like programming in java.
# Then replace all occurance of "java" with "python" in above line.

# Step 1: Create the file and write initial content
with open("/kaggle/working/practice.txt", "w") as f:
    f.write("Hi everyone \nwe are learning file i/o \nusing java \nI like programming in java.")

# Step 2: Read the file, replace "java" with "python", and write back
with open("/kaggle/working/practice.txt", "r") as f:
    content = f.read()

content = content.replace("java", "python")  # replace all occurrences

with open("/kaggle/working/practice.txt", "w") as f:
    f.write(content)

# Step 3: Read the final file to display content
with open("/kaggle/working/practice.txt", "r") as f:
    final_content = f.read()

print("Final file content:\n")
print(final_content)

# Step 4: Check if the word "learning" exists
if "learning" in final_content:
    print("\nThe word 'learning' exists in the file.")
else:
    print("\nThe word 'learning' does NOT exist in the file.")
```
