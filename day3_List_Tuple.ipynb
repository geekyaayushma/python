```python
marks = [ 87 , 64 , 33 ,95 , 76]

print(marks [:])
print (marks [3:4]) #ending index is not included
```

# List Methods

* list.append() - adds one element at the end
* list.sort() - sorts in ascending order
* list.sort( reverse=True ) - sorts in descending order
* list.reverse() - reverse list
* list.insert( idx,el) - insert element at index
* list.remove() - removes the specified item
* list.pop(idx) - remove element in index

```python
list = [2,1,3]

list.append(4)

print(list)

print (list.append(5)) #doesnot return anything
```

```python
list = [33 ,1 , 44 ,33]
name = ["dif" ,"Abb" , "Cla"]

list.sort()
name.sort()

print(list)
print(name)
```

```python
list = [33 ,1 , 44 ,33]
fruits = ["banana" , "apple" ,"Avocado"]

list.sort(reverse= True)
fruits.sort(reverse= True)

print(list)
print(fruits)
```

```python
fruits = ["banana" , "apple" ,"avocado"]

fruits.reverse()

print(fruits)
```

```python
fruits = ["banana" , "apple" ,"avocado"]

fruits.insert (0 , "mango")

print(fruits)
```

```python
thislist = ["apple", "banana", "cherry", "banana", "kiwi"]

thislist.remove("banana")

print(thislist)
```

```python
thislist = ["apple", "banana", "cherry", "banana", "kiwi"]

thislist.pop(0)

print(thislist)
```

# Tuple Methods
* tup.index(el) - returns index of first occurance
* tup.count(el) - counts total occurance

```python
tup = (1 ,3 , 5 ,6)

num1 = tup.index(1)

print(num1)
```

```python
tup = (1 ,3 , 5 ,6 ,1)

count = tup.count(1)

print(count)
```
