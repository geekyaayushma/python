# While loop

```python
#while loop

count = 1 #iterator
 
while count <= 5: #iteration
    print("Count is:", count)
    count += 1
```

```python
# Break and continue demo

i = 30

while i <= 300:
    if i == 90:
        i += 30     # increment BEFORE continue to avoid infinite loop
        continue    # skip printing 90
    if i == 210:    # example stop condition
        break       # stop completely
    print(i)
    i += 30

```

```python
# Print numbers from 1 to 100

i = 1
while i <= 100:
    print(i, end=" ")
    i += 1
```

```python
# Print the elements of the following list using a loop : [1,4,9,16,25,36,49,64,81,100]

num = [1,4,9,16,25,36,49,64,81,100]
i = 0

while i < len(num):
    print("The list number at index" , i , "is" , num[i])
    i +=1
```

```python
# Search for a number x in this tuple using loop : [1,4,9,16,25,36,49,64,81,100]

tup = (1, 4, 9, 16, 25, 36, 49, 64, 81, 100)

x = int(input("Enter a number: "))  # convert to int
i = 0

while i < len(tup):
    if tup[i] == x:
        print("The given number is", tup[i])
        break   
    i += 1
else:
    print("Number not found in tuple")
```

# for loop

```python
veggies = ["potato" ,"tomato" ,"cucumber"]

for val in veggies:
    print(val)
```

```python
tup = (1,2,4,5,4,2)

for num in tup :
    print(num)
```

```python
str = "Python"

for char in str:
    if char == 'o':
        print ("o found")
        break   
    print(char)     
else :
        print("Not found")
    
```

range()

Range functions returns a sequence of numbers, starting from 0 by defult, and increments by 1 (by defult) and stops before a specified number.

syntax:

range(start?,stop,step?)

```python
for i in range(5):
    print(i , end='\n')
for i in range(1,5):
    print(i , end='\n')
for i in range (1,10,3):
    print(i, end='\n')
```

```python
#print odd
print("odd num:")
for i in range (1 , 100 , 2):
    print( i , end='\t')
print("\n")

#print even
print("even num:")
for i in range (2 , 100 ,2):
    print( i , end ='\t')
```

```python
# WAP to find the sum of first n numbers. (using while)

n = int(input("Enter a number: "))
i = 1
sum = 0

while i <= n:
    sum += i
    i += 1

print("The sum of first", n, "numbers is:", sum)

```

```python
# WAP to find the factorial of first n numbers.(using for)

n = int(input("Enter a number: "))

factorial = 1 

for i in range(1, n + 1):  
    factorial *= i          
print("Factorial of", n, "is:", factorial)
```
