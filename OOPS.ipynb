# Object Oriented Programming



**Class and Object :**

A **class** is a blueprint or a template for creating objects. It defines the structure and behavior that all objects of that class will possess.

An **object** is an instance of a class. It is a concrete entity created based on the blueprint provided by the class.

```python
class Student:
    name = "Aayushma Mahat",
    roll_no = 22

# creating object (instance)

s1 = Student()
s2 = Student()
print(s1)
print(s1.name)
print(s2.name)
```

```python
class Car:
    color = "Blue",
    brand = "Mercedes"

car1 = Car()
print(car1.color)
print(car1.brand)
```


**__init__function / Constructor**


All classes have a function called __init__(), which is always executed when the object is being initied.


```python
#creating class

class Car:
    color = "Blue",

    #defult_constructors    
    def __init__():
        pass

    #parametarized constructor 
    def __init__(self , fullBrand , id):
        self.brand = fullBrand
        self.id = id
        print("adding new car in database",self.brand)

#creating object

car1 = Car("honda" , 1)
print(car1.brand , car1.id)

car2 = Car("BYD" , 2)
print(car2.brand , car2.id)
```

The **self parameter** is a reference to the current instance of the class, and is used to acess variables that belongs to the class.

```python
class Student:
    # Class attribute - shared by all instances
    school = "ABC High School"
    name = "anonymos" # class atr
    
    def __init__(self, name = "Aayu", roll_no = "1"):
        # Object attributes - unique to each instance
        self.name = name #obj attr > class attr 
        self.roll_no = roll_no
        print(f"Student {self.name} added to database")

# Creating objects with default values
s1 = Student()

# Creating objects with custom values
s2 = Student("Ram Sharma", 25)
s3 = Student("Sita Poudel", 30)

print(f"s1 name: {s1.name}")
print(f"s2 name: {s2.name}")
print(f"s3 name: {s3.name}")

print(f"school (class attr): {Student.school}")
```

```python
# Create student class that takes name & marks of 3 subjects as arguments in constructor. Then create a method to print the average.

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def print_average(self):
        total = 0
        for mark in self.marks:  
            total += mark
        average = total / len(self.marks)
        print(f"{self.name}'s average marks: {average:.2f}")

student1 = Student("Alice", [85, 92, 78])
student2 = Student("Bob", [76, 88, 91])

student1.print_average()
student2.print_average()


```

**Static Methods :**

Methods that don't use the self parameter (works at class level).

syntax:
@ static method
..... #decorator

Decorator allows us to wrap another function in order to extend the behaviour of the wrapped function, without premanently modifying it.

```python
class Student:
    def __init__(self, name ):
        self.name = name

    @staticmethod #decorator
    def hello():
        print("Konichiwa")

s1 = Student("Nisha")

s1.hello()
```

```python
# Create Account class with 2 attributes -balance and account no. Create methods for debit, credit and printing the balance.

class Account:
    def __init__(self, balance, account_no):
        self.balance = balance
        self.account_no = account_no

    def debit(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            print(f"Debited {amount}. Remaining balance: {self.balance}")
        else:
            print("Insufficient balance!")

    def credit(self, amount):
        self.balance += amount
        print(f"Credited {amount}. New balance: {self.balance}")

    def print_balance(self):
        print(f"Account {self.account_no} balance: {self.balance}")

Acc1 = Account(10000, 323289328430)

Acc1.debit(2000)       
Acc1.credit(5000)      
Acc1.print_balance()   
     
```

**Inheritance:**

When one class(child/derived) derives the properties and method of another class(parent/base).

Types of inheritance :
* Single Inheritance
* Multi-level Inheritance
* Multiple Inheritance

```python
#Single Inheritance
class Car:
    
    @staticmethod
    def start():
        print("car start")
        
    @staticmethod
    def stop():
        print("car stop")

class ToyotaCar(Car):
    def __init__(self,name):
        self.name = name

car1 = ToyotaCar("Fortuner")
car2 = ToyotaCar("Pirus")

print(car1.start())
```

```python
#Multi-level Inheritance

class Car:

    @staticmethod
    def start():
        print("Start...")

    @staticmethod
    def stop():
        print("Stop...")

class Toyota(Car):
    def __init__(self,brand):
        self.brand = brand

class Fortuner(Toyota):
    def __init__(self,type):
        self.type = type

car1 = Fortuner("Diseal")
car1.start()
```

```python
 #Multiple Inheritance

class A:
    varA = "Welcome to class A"

class B:
    varB = "Welcome to class B"

class C(A,B):
    varC = "Welcome to class C"

abc1 = C()
print(abc1.varA , abc1.varB , abc1.varC)
```

**Super Method:**

super() method is used to access methods of the parent class.

```python
#Multi-level Inheritance

class Car:
    def __init__(self,type):
        self.type = type 
        
    @staticmethod         
    def start():
        print("Start...")

    @staticmethod
    def stop():
        print("Stop...")

class Toyota(Car):
    def __init__(self,name,type):
        self.name = name
        super().__init__(type)
        super().start()

car1 = Toyota("Fortuner","electric")
print(car1.type)
```

**Class Method**

A class method is boubd to the class anad receives the class as an implicit first argument.
Note - static method can't access or modify class state and generally for utility.
syntax: @classmethod #decorator


```python
class Person:
    name = "anomymous"

    def changeName(self,name):
        #this
        self.__class__.name = name
        #or this Person.name = name 

p1 = Person()
p1.changeName("Aayushma")
print(p1.name)
print(Person.name)
```

```python
class Person:
    name = "anomymous"
    
    @classmethod
    def changeName(cls,name):
        cls.name = name
 

p1 = Person()
p1.changeName("Aayushma")
print(p1.name)
```
