# Conditional statement



```python
""" syntax :
if-elif-else

if (condition) : 
   statement1;
 elif (condition):
     statement2;
 else (condition):
     statementN;    """

age = 21

if age >= 20:
    print("can drink")
    print("can vote")
elif age >= 18:
    print("can vote")
else :
    print("cannot vote")
    print("cannot drink")
```

```python
marks = 89

if (marks >= 90):
   grade = "A"
elif (marks >= 80 and marks< 90):
   grade = "B"
elif (marks >= 70 and marks< 80):
   grade = "C"
else:
   grade = "D"

print ("The grade is" , grade)
```

```python
#Nesting conditional satatements

age = 90

if (age >=18 ):
    if(age >=80) :
        print ("Cannot drive")
    else :
        print ("Can drive")
else :
    print ("Cannot drive")
```
