# Dictionary methods

* myDict.keys() - returns all the keys
* myDict.values() - returns all the values
* myDict.items() - returns all (key,val) pairs as tuples
* myDict.get("key"")  - returns the key according to values
* myDict.update(newDict) - insert the specific items to the dictionary

```python
info = {
    "name" : "aayushma",
    "subjects":{
        "phy" : 26,
        "chem" : 30
    }
}

print(info.keys())
print(list(info.keys()))
print(len(info.keys()))
```

```python
info = {
    "name" : "aayushma",
    "subjects":{
        "phy" : 26,
        "chem" : 30
    }
}

print(info.values())
print(list(info.values()))
print(len(info.values()))
```

```python
info = {
    "name" : "aayushma",
    "subjects":{
        "phy" : 26,
        "chem" : 30
    }
}

print(info.items())
pairs =list(info.items())
print(pairs[0])
```

```python
info = {
    "name" : "aayushma",
    "subjects":{
        "phy" : 26,
        "chem" : 30
    }
}

# print(info("surname")) produce err
print(info.get("surname"))
```

```python
info = {
    "name" : "aayushma",
    "subjects":{
        "phy" : 26,
        "chem" : 30
    }
}

print(info)

new_dict = { "name" : "aayu", 
            "age" : 21 }

info.update(new_dict)

print(info)
```

# Set method

* set.add(el) - add elements to set
* set.remove(el) - remove element from set
* set.clear() - empties the set
* set.pop() - removes the random values
* set.union(set2) - combines both set and return values
* set.intersection(set2) - combines common values and return values

```python
collection = set()

collection.add(1)
collection.add("hi")
collection.add(1)
collection.add((2,34,5))

print(collection)
```

```python
collection = {2,3,4,5,"hi"}

collection.remove("hi")

print(collection)
```

```python
colection = {2,3,4,5,"hi"}

collection.clear()

print(collection)
```

```python
collection = { 2 ,45 ,55}

collection.pop()
collection.pop()
print(collection)
```

```python
set1 = { "Apple" , "Banana"}
set2 = {"cucumber" , "radish"}

set1.union(set2)

print(set1)
```

```python
set1 = { "Apple" , "Banana" , "cucumber"}
set2 = {"cucumber" , "radish"}

set1.intersection(set2)

print(set1)
```
